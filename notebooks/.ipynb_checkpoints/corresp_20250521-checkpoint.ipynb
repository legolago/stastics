## コレスポンデンス分析とは
### 複雑な表データを見やすい地図のように変換する方法ともいえる。例えば、様々なファッションブランドと「豪華な」「若々しい」などのイメージの関係を、一目で理解できる「位置関係の地図」に変えるもの。
### コレスポンデンス分析は複雑な情報を「空間上の位置関係」に置き換えることで、直感的に理解できるよう視覚化するツール。

## 手順
### 1.データの準備: 行（例：イメージ属性）と列（例：ブランド）のクロス表を用意
### 2.関連性の計算: どのイメージとどのブランドの結びつきが強いかを計算
### 3.次元の削減: 複雑な関係を2次元（x軸とy軸）に圧縮
### 4.次元表示: 関連の強いものが近く、弱いものが遠くなるように配置します

## コレスポンデンス分析の効果
### 1.複雑なデータを視覚化できる
- 数字の羅列から何かを読み取るのは難しいが、図を見れば一目で傾向がわかる。例えば、「シャネルは豪華で高級なイメージに近い」「ユニクロはシンプルで親しみやすいイメージに近い」などの関係が視覚的に理解でる。
### 2.隠れた関係性を発見できる
- 一見して関係がないように見えるものでも、分析すると「実は似ている」ということがわかることがある。例えば、異なるブランドが実は同じような消費者イメージを持っていることなどが見えてきます。
### 3.市場ポジショニングの把握
- ファッションブランドの例では、各ブランドが消費者の心の中でどのような位置にあるのかを理解できます。これは商品開発やマーケティング戦略を考える上で非常に役立つ。
### 4.データの要約
- 膨大なデータを少ない情報で表現できるため、全体像を素早く把握できる。

## バイプロットの基本的な考え方
### バイプロットとは二種類の異なるデータ（行と列）を同じ空間に描画する方法。本来異なる種類のデータ（例：ブランドとイメージ特性）を同じ座標系で表現するというやや不思議な概念ため難しく感じてしまう。
### 1.共通の「関係性空間」を作る
- コレスポンデンス分析では、行と列の関係を「期待値からのずれ（残差）」として計算。このずれが大きいほど、その行と列の組み合わせは「特徴的」と言える。
### 2.この残差を使って「行」も「列」も同じ空間に投影。
### 3.原点の意味
- バイプロットの中心（原点）は「平均的な関係性」を表します。原点から離れるほど、そのブランドやイメージは「特徴的」であることを意する。
### 4.解釈の仕方
- 近い位置にある行と列: 強い関連がある（例：「シャネル」と「豪華な」が近ければ、シャネルは豪華なイメージが強い）
- 同じ方向にある行/列: 似たようなプロファイルを持つ（例：「シャネル」と「ルイヴィトン」が近ければ、似たようなイメージプロファイルを持つ）
- 反対方向にある行/列: 対照的なプロファイルを持つ

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import matplotlib.font_manager as fm
import matplotlib as mpl
import japanize_matplotlib  


# 日本語フォントの設定
#plt.rcParams['font.family'] = 'IPAexGothic'  # または他の日本語対応フォント（'MS Gothic', 'Yu Gothic'など）

# CSVファイルを読み込む
df = pd.read_csv('fashion_brands.csv')

# 'image'列をインデックスに設定
df.set_index('image', inplace=True)

# コレスポンデンス分析のための関数
def correspondence_analysis(df, n_components=2):
    # コンティンジェンシー表（クロス集計表）を正規化
    row_sum = df.sum(axis=1)
    col_sum = df.sum(axis=0)
    total = row_sum.sum()
    
    # 期待値の計算
    expected = np.outer(row_sum, col_sum) / total
    
    # カイ二乗距離の計算
    chi2_dist = ((df.values - expected)**2 / expected).sum().sum()
    
    # 行と列の質量（重み）
    row_mass = row_sum / total
    col_mass = col_sum / total
    
    # 標準化残差行列
    std_residuals = (df.values - expected) / np.sqrt(expected)
    
    # 特異値分解（SVD）のための行列
    row_std = std_residuals / np.sqrt(row_mass.values.reshape(-1, 1))
    col_std = std_residuals.T / np.sqrt(col_mass.values.reshape(-1, 1))
    
    # SVDの計算
    u, s, vt = np.linalg.svd(std_residuals, full_matrices=False)
    
    # 主成分の変換
    row_coords = row_std @ vt.T[:, :n_components]
    col_coords = col_std @ u[:, :n_components]
    
    # 慣性（寄与率）の計算
    inertia = s**2 / np.sum(s**2)
    
    return {
        'row_coords': row_coords,
        'col_coords': col_coords,
        'eigenvalues': s**2,
        'inertia': inertia,
        'chi2': chi2_dist,
        'row_mass': row_mass,
        'col_mass': col_mass
    }

# コレスポンデンス分析の実行
ca_results = correspondence_analysis(df)

# 結果の出力
row_coords = ca_results['row_coords']
col_coords = ca_results['col_coords']
inertia = ca_results['inertia']

# 結果のデータフレーム化
row_df = pd.DataFrame(row_coords, index=df.index, columns=[f'Dim {i+1}' for i in range(row_coords.shape[1])])
col_df = pd.DataFrame(col_coords, index=df.columns, columns=[f'Dim {i+1}' for i in range(col_coords.shape[1])])

# 寄与率の出力
print("各次元の寄与率:")
for i, contrib in enumerate(inertia[:2]):
    print(f"第{i+1}次元: {contrib:.2%}")

# プロット
plt.figure(figsize=(12, 10))

# 行ポイント（イメージ）のプロット
plt.scatter(row_coords[:, 0], row_coords[:, 1], color='blue', label='イメージ')
for i, (idx, row) in enumerate(row_df.iterrows()):
    plt.annotate(idx, (row_coords[i, 0], row_coords[i, 1]), color='blue', fontsize=11)

# 列ポイント（ブランド）のプロット
plt.scatter(col_coords[:, 0], col_coords[:, 1], color='red', label='ブランド')
for i, (idx, row) in enumerate(col_df.iterrows()):
    plt.annotate(idx, (col_coords[i, 0], col_coords[i, 1]), color='red', fontsize=11)

# プロットの設定
plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='gray', linestyle='-', alpha=0.3)
plt.grid(alpha=0.3)
plt.xlabel(f'第1次元 ({inertia[0]:.2%})', fontsize=12)
plt.ylabel(f'第2次元 ({inertia[1]:.2%})', fontsize=12)
plt.title('ファッションブランドとイメージのコレスポンデンス分析', fontsize=14)
plt.legend()

# グラフの保存
plt.tight_layout()
plt.savefig('brand_image_correspondence_analysis.png', dpi=300)
plt.show()

# scikit-learnを使用したコレスポンデンス分析の実装（代替方法）
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

def ca_with_sklearn(df):
    # データの前処理
    row_sum = df.sum(axis=1)
    col_sum = df.sum(axis=0)
    total = row_sum.sum()
    
    # 行と列の質量（重み）
    row_mass = row_sum / total
    col_mass = col_sum / total
    
    # コレスポンデンス行列の計算
    p = df / total
    
    # 行と列の質量の平方根の逆数
    row_mass_sqrt_inv = np.diag(1.0 / np.sqrt(row_mass.values))
    col_mass_sqrt_inv = np.diag(1.0 / np.sqrt(col_mass.values))
    
    # 標準化されたコレスポンデンス行列
    s = row_mass_sqrt_inv @ (p.values - np.outer(row_mass, col_mass)) @ col_mass_sqrt_inv
    
    # PCAを使用した分解
    pca = PCA()
    pca.fit(s)
    
    # 座標の計算
    row_coords = row_mass_sqrt_inv @ pca.components_.T[:, :2]
    col_coords = col_mass_sqrt_inv @ pca.components_[:2, :].T
    
    return {
        'row_coords': row_coords,
        'col_coords': col_coords,
        'inertia': pca.explained_variance_ratio_
    }

# scikit-learnを使用した代替実装の実行とプロット（オプション）
# ca_sklearn_results = ca_with_sklearn(df)
# ここに追加のプロットコードを記述可能
```
